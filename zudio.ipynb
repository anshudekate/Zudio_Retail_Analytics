```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv(r"D:\Data anaytics\Zudio_sales_data.csv")

# Calculate Pearson correlation between Quantity and Sales_Profit
corr, p_value = pearsonr(df['Quantity'], df['Sales_Profit'])

print(f"Correlation coefficient: {corr:.3f}")
print(f"P-value: {p_value:.3g}")

# Visualize the relationship
plt.scatter(df['Quantity'], df['Sales_Profit'])
plt.xlabel('Quantity')
plt.ylabel('Sales Profit')
plt.title('Correlation between Quantity and Sales Profit')
plt.show()
```
